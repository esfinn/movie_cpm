```python
import numpy as np
import scipy as sp
from matplotlib import pyplot as plt
import pandas as pd
import h5py
import seaborn as sns
```

```python
top_dir = "../data/"
list_of_dfs = ['unres_behav_data.csv', 'res_behav_data.csv', 'emot_pc_scores.csv', 'cogn_pc_scores.csv', 'motion_data.csv', 'tod_data.csv']
```

```python
for d,df in enumerate(list_of_dfs):
    tmp = pd.read_csv(top_dir+df, dtype={'Subject': 'str'})
    tmp.set_index("Subject", inplace=True)
    print('Doing df {}...number of columns in this df: {}'.format(df, tmp.shape[1]))
    
    if d==0:
        behav_data = tmp
    else:
        # behav_data = behav_data.join(tmp, how='outer')
        behav_data = pd.concat([behav_data, tmp], axis=1)
```
