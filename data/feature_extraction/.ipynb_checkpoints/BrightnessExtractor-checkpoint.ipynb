```python
import imageio
imageio.plugins.ffmpeg.download()
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import pliers
%matplotlib inline

import os
```

```python
from pliers.extractors import BrightnessExtractor
from pliers.stimuli import VideoStim
video = VideoStim('/Users/finnes/Documents/HCP_7T/7T_movie_stimulus/Post_20140821_version/7T_MOVIE1_CC1_v2.mp4')
bright_ext = BrightnessExtractor()
bright_result = bright_ext.transform(video)
```

```python
result_df = [f.to_df() for f in bright_result]

result_df = pd.concat(result_df)

result_df.columns

result_df
```
